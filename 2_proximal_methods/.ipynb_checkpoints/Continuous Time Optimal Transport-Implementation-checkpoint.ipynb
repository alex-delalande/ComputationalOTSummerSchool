# Benamou and Brenier formulation of optimal transport using proximal optimization methods. 
This notebook is based on *Optimal Transport with Proximal Splitting*, Papadakis, Peyré and Oudet.

```python
import numpy as np
import scipy as sp
import pylab as pl
from scipy.fftpack import dct,idct
```

We start with a 1D implementation of the method. The initial and final densities are vectors of length $nx$ and we use $nt$ timesteps.

```python
# initialize these parameters
nx, nt = 100, 8
# nx points equally distributed in [0,1]
x = np.linspace(-0.0, 1.0, nx)
# initialize your initial and final densities
gauss = lambda x, sigma, mean: np.exp(-((x-mean)/sigma)**2)/np.sum(np.exp(-((x-mean)/sigma)**2))
initialDensity, finalDensity = gauss(x,0.13,0.25), gauss(x,0.08,0.8)
pl.plot(x,finalDensity), pl.plot(x,initialDensity), pl.show();
```

We define the shape of the staggered grids and the initialized variables.

```python
shapeStag_x = (nx + 1, nt) # shape of the staggered grid with respect to x
shapeStag_t = (nx, nt + 1) # shape of the staggered grid with respect to t
shapeCentered = (nx, nt)   # shape of the centered grid
### actually more efficient in python to consider the transpose since nx >> nt.
### initial variables on the staggered and centered grid
staggeredMomentum = np.zeros(shapeStag_x)
density = np.zeros(shapeCentered) 
momentum = np.zeros(shapeCentered)
staggeredDensity = np.zeros(shapeStag_t)
staggeredDensity[:,0] = initialDensity
staggeredDensity[:,-1] = finalDensity
### time linear interpolation of densities
for i in range(nt + 1):
    temp = float(i)/(nt + 1.)
    staggeredDensity[:,i] = temp *finalDensity + (1.0-temp)*initialDensity
for i in range(nt):
    temp = float(i)/(nt + 1.)
    density[:,i] =temp *finalDensity + (1.0-temp)*initialDensity  
### definition of the boundary conditions
b0 = np.concatenate((np.copy(initialDensity),np.copy(finalDensity),np.zeros(nt),np.zeros(nt)))
```

Time linear interpolation of densities

```python
for i in range(nt+1):
    pl.plot(x, staggeredDensity[:,i], label="t={}".format(i))
pl.legend()
pl.show()
```

 The discrete optimization problem is the following: $U$ is the vector on the staggered grid and $V$ is the vector on the centered grid. $C$ denotes the divergence free constraint and the boundary conditions (in time and space). Last, $I$ denotes the interpolation constraint that links $U$ and $V$. $F$ represents the kinetic energy on the centered grid.
 \begin{equation} \min_{U,V} F(V) + \iota_{C}(U) + \iota_{I}(U,V) \end{equation} 
The application of the Douglas-Rachford algorithm requires the definition of two functions for which the proximal operator can be computed "easily". One can choose
 \begin{equation} \begin{cases} G_1(U,V) = F(V) + \iota_{C}(U) \\ G_2(U,V) =  \iota_{I}(U,V)\,.\end{cases}\end{equation} 
Then, the proximal operator of $G_1$ consists in computation of the proximal operator of $F$ and $\iota_C$ since they apply independently to $U$ and $V$. 

### Step 1: Computation of the proximal map for $F$.
The proximal operator for $F$ requires the computation of a third order polynomial function at each point of the centered grid.

```python
# Computation of the proximal operator of F
# For that one needs the highest root of a third order polynomial function that can be computed as follows:
def computeRoot3rdPolynomial(l):
    a,b,c,d = l[0],l[1],l[2],l[3]
    p = -((b/a)**2)/3 + c/a
    q = (2.*(b/a)**3)/27 - b*c/(a**2)/3 + d/a
    delta = q**2 + 4.0/27.0*(p**3)
    if delta > 0:
        u = ((-q + np.sqrt(delta))/2.)**(1./3.)
        if (-q - np.sqrt(delta))/2.<0:
            v = -((q + np.sqrt(delta))/2.)**(1./3.)
        else:
            v = ((-q - np.sqrt(delta))/2.)**(1./3.)
        z = u + v - (b/ a) /3.
    elif delta < 0:
        u = (np.complex(-q , np.sqrt(-delta))/2.)**(1./3.)
        z = (u + np.complex.conjugate(u) - (b/a) /3.).real
    else:
        z =  (3*q / p - (b/ a) /3.)
    return z

def evaluatePol(a,b,c,d,x):
    return a*(x**3) + b*(x**2) + c*(x) + d
# if needed, check that the first function gives the correct result.

def proximalKineticEnergy(density,momentum,alpha):
    """
    The parameter alpha is instantiated in DR optimization function and is common to proximal operators. 
    (However, since all the other proximal operators are projections, 
    these proximal operators are independent of alpha.)
    """
    temp = np.zeros(shapeCentered + (4,))
    temp[:,:,0 ] = 1
    temp[:,:,1 ] = 4. * alpha - density
    temp[:,:,2 ] = 4.*alpha*(alpha-density)
    temp[:,:,3 ] = -alpha * momentum**2 - 4.*(alpha**2)*density
    new_dens = np.apply_along_axis(computeRoot3rdPolynomial,-1,temp)
    new_dens = np.maximum(new_dens,np.zeros(shapeCentered))
    new_mom = momentum * new_dens/(new_dens + 2*alpha)
    return new_dens,new_mom 
```

### Step 2: Computation of the proximal map for $\iota_C$.
The proximal operator for $\iota_C$ can be efficiently computed via FFT methods since it requires to solve a Poisson equation with Neumann boundary conditions. It can be done using the discrete cosine transform (DCT).

```python
### helper for the numerical solution of the Poisson equation in Fourier.
matrixPoisson = np.zeros(shapeCentered)
for i in range(shapeCentered[0]):
    for j in range(shapeCentered[1]):
        temp = 2 * (1 - np.cos(np.pi * float(i)/(shapeCentered[0]))) 
        temp += 2 * (1 - np.cos(np.pi * float(j)/(shapeCentered[1])))
        if temp==0:
            temp=1e-10 # put a small positive value instead of 0 to avoid overflow
        matrixPoisson[i,j] = 1.0/(temp)

def proximalProjectionOnDivFreePoisson2(staggered_density,staggered_momentum):
    """
    Divergence operator by finite differences
    """
    projection = projectionOnBC(staggered_density,staggered_momentum,b0)
    div_proj = divergenceOperator(*projection)
    temp = solvePoisson(div_proj)
    temp_bis = adjointDivergenceOperator(temp)
    temp_bis2 = projectionOnBC(*temp_bis,b0 * 0)
    result0 = projection[0] - temp_bis2[0]
    result2 = projection[1] - temp_bis2[1]
    return result0,result2

def solvePoisson(centered_grid):
    """
    Solve Poisson equation using dct
    """
    fourier = dct2D(centered_grid)
    fourier_sol = np.multiply(matrixPoisson,fourier)
    result = idct2D(fourier_sol)
    return result 

def dct2D(grid):
    temp = dct(grid,type=2,norm = "ortho")
    temp2 = dct(temp.transpose(),type=2,norm = "ortho")
    return temp2.transpose()

def idct2D(grid):
    temp = idct(grid,type=2,norm = "ortho")
    temp2 = idct(temp.transpose(),type=2,norm = "ortho")
    return temp2.transpose()

def divergenceOperator(staggered_density,staggered_momentum,dim = None):
    """
    Divergence operator by finite differences
    """
    div_density = finiteDifference(staggered_density,direction = 1)
    div_momentum = finiteDifference(staggered_momentum,direction = 0)
    return div_density+div_momentum    

def adjointDivergenceOperator(centered_grid):
    stag_dens = adjointFiniteDifference(centered_grid,1)
    stag_mom = adjointFiniteDifference(centered_grid,0)
    return stag_dens,stag_mom

def finiteDifference(array,direction=0):
    """
    Finite difference, direction is the index representing space (index 0) or time (index 1)
    """
    if direction==0:
        result = array[1:,:] - array[:-1,:]
    elif direction==1:
        result = array[:,1:] - array[:,:-1]
    else:
        return "direction must be 0 or 1"
    return result

def adjointFiniteDifference(arrayTemp,direction):
    if direction ==0:
        result = np.zeros_like(staggeredMomentum)
        result[1:,:] = arrayTemp
        result[:-1,:] -= arrayTemp
    if direction == 1:
        result = np.zeros_like(staggeredDensity)
        result[:,1:] = arrayTemp
        result[:,:-1] -=arrayTemp
    return result 

def projectionOnBC(stag_d,stag_m,b0):
    """
    Projection on the boundary constraint: put the values of the boundary to b0
    """
    stag_dens_temp,stag_mom_temp = np.copy(stag_d),np.copy(stag_m)
    indice = nx
    yb = b0
    stag_dens_temp[:,0] = yb[:indice]
    stag_dens_temp[:,-1] = yb[indice:2*indice]
    indice += indice
    stag_mom_temp[0,:] = yb[indice:indice + nt]
    indice += nt
    stag_mom_temp[-1,:] = yb[indice:indice + nt]
    return stag_dens_temp,stag_mom_temp
```

### Step 3: Computation of the proximal map for $\iota_I$.
The proximal operator for $\iota_I$ can be computed using the $LU$ decomposition of the operator. But, to do it a simple way, just implement the projection by storing the explicit inverse of the matrix that operates on each dimension.

```python
def interpolateFromStaggeredToGrid(staggered,dim=0):
    ### TODO: implement the interpolation operator that maps a staggered grid onto a centered grid 
    ### along a particular dimension
    if dim==0:
        result = (staggered[1:,:] + staggered[:-1,:])*0.5
    elif dim ==1:
        result  =  (staggered[:,1:] + staggered[:,:-1])*0.5
    return result

def transposeInterpolation(grid,dim=0):
    ### TODO: implement the transpose of the interpolation operator that maps a staggered grid 
    ### onto a centered grid along a particular dimension
    liste = list(np.shape(grid))
    if dim==0:
        liste[0]+=1
        result = np.zeros(liste)
        result[1:,:] += grid
        result[:-1,:] += grid
    if dim==1:
        liste[1]+=1
        result = np.zeros(liste)
        result[:,1:] += grid
        result[:,:-1] += grid
    return 0.5*result

def buildProjectionInvertibleOperator(dim,operatorProjection):
    if dim ==0:
        dimension = shapeStag_x[dim]
        matrix = buildMatrix(interpolateFromStaggeredToGrid,dimension)
        matrix2 = buildMatrix(transposeInterpolation,dimension-1)
        temp = np.dot(matrix2,matrix)
        result = temp + np.eye(np.shape(temp)[0])
        operatorProjection[dim] = np.linalg.inv(result)
    elif dim==1:
        dimension = shapeStag_t[dim]
        matrix = buildMatrix(interpolateFromStaggeredToGrid,dimension)
        matrix2 = buildMatrix(transposeInterpolation,dimension-1)
        temp = np.dot(matrix2,matrix)
        result = temp + np.eye(np.shape(temp)[0])
        operatorProjection[dim] = np.linalg.inv(result)
    return None

def buildMatrix(linearMap,dim):
    inputs = np.random.rand(dim,1)
    outputs = linearMap(inputs)
    result = np.zeros((np.shape(outputs)[0],dim))
    for i in range(dim):
        x = np.zeros((dim,1))
        x[i]=1
        result[:,i] = linearMap(x)[:,0]
    return result 




operatorProjection = [None,None]
buildProjectionInvertibleOperator(0,operatorProjection)
buildProjectionInvertibleOperator(1,operatorProjection)


def proximalProjectionOfInterpolation(staggered,onTheGrid,dim):
    temp_vector = staggered + transposeInterpolation(onTheGrid,dim)
    if dim==0:
        temp2 = np.dot(operatorProjection[dim],temp_vector)
    elif dim==1:
        temp = np.dot(operatorProjection[dim],np.transpose(temp_vector))
        temp2 = np.transpose(temp)
    return temp2,interpolateFromStaggeredToGrid(temp2,dim)
```

### Step 4: Implementation of the Douglas-Rachford algorithm.

```python
def proximalG1(staggeredDensity,staggeredMomentum,density,momentum,alpha):
    # attention, addition de tuple est un tuple
    return proximalProjectionOnDivFreePoisson2(staggeredDensity,staggeredMomentum) + \
           proximalKineticEnergy(density, momentum,alpha)

def proximalG2(staggeredDensity,staggeredMomentum,density,momentum):
    new_stag_dens,new_dens = proximalProjectionOfInterpolation(staggeredDensity,density,1)
    new_stag_mom,new_mom = proximalProjectionOfInterpolation(staggeredMomentum,momentum,0)
    return new_stag_dens,new_stag_mom,new_dens,new_mom

def optimizationDR(iterations,staggeredDensity,staggeredMomentum,density,momentum):
    ### choice of parameters for DR
    beta = 1.8
    alpha = 0.05
    ### linear combination of tuples
    average1 = lambda z: 2*z[0] - z[1]
    average2 = lambda z: z[0] - z[1]
    average3 = lambda z: beta * z[0] + z[1]
    ### initialization
    w = (staggeredDensity, staggeredMomentum, density, momentum)
    z = (staggeredDensity, staggeredMomentum, density, momentum)
    ### DR iterations
    for i in range(iterations):
        temp0 = tuple(map(average1,zip(z,w)))
        temp = proximalG2(*temp0)
        temp3 = tuple(map(average2,zip(temp,z)))
        w = tuple(map(average3,zip(temp3,w)))
        z = proximalG1(*w,alpha)
    return z
```

```python
iterations = 1000
staggeredDensityResult,staggeredMomentumResult,densityResult,momentumResult = optimizationDR(1000,staggeredDensity,staggeredMomentum,density,momentum)
```
